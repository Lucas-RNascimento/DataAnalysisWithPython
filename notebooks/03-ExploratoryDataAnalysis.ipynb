### Variaveis Numericas Continuas


Variáveis numéricas contínuas são aquelas que podem assumir qualquer valor dentro de um determinado intervalo. Elas podem ser do tipo `'int64'` ou `'float64'`. Uma ótima maneira de visualizar essas variáveis é por meio de gráficos de dispersão (_scatterplots_) com linhas ajustadas.

Para começar a entender a relação (linear) entre uma variável individual e o preço, podemos utilizar **"regplot"**, que plota o gráfico de dispersão e adiciona a linha de regressão ajustada aos dados. Isso será útil posteriormente para visualizar o ajuste do modelo de regressão linear simples. 



````python
## Positive Linear Relationship

sns.regplot(x='engine-size', y='price', data=df)
plt.ylim(0,)


# Esse trecho `plt.ylim(0,)` está ajustando os limites do eixo Y do gráfico. 

#- `plt.ylim(0,)` está definindo o limite inferior do eixo Y para **0**.  

#- O argumento `0,` com a vírgula indica que só o limite inferior está sendo especificado, 
# e o limite superior será ajustado automaticamente pelo Matplotlib, com base nos dados do 
# gráfico.  

#Ou seja, ao adicionar essa linha, você está garantindo que o eixo Y nunca terá valores 
# negativos, o que pode ser útil quando se trata de preços, já que eles não podem ser 
# negativos.  
````

### **Análise Estatística Descritiva**

A função `describe()` do Pandas calcula automaticamente estatísticas básicas para todas as variáveis **contínuas**. Qualquer valor **NaN** (faltante) é automaticamente ignorado nesses cálculos.

#### **Métricas geradas pela função `describe()`:**
- **Contagem (`count`)** – Número total de valores não nulos na variável
- **Média (`mean`)** – Valor médio dos dados
- **Desvio padrão (`std`)** – Medida da dispersão dos valores
- **Mínimo (`min`)** – Menor valor encontrado na variável
- **Quartis (`25%`, `50%` - mediana, `75%`)** – Distribuição dos dados em percentis
- **Máximo (`max`)** – Maior valor da variável

Essa função é essencial para obter uma visão rápida sobre a distribuição dos dados antes de realizar análises mais complexas, como testes estatísticos ou construção de modelos preditivos. 


````python
df.describe()
````

````python
# We can apply the method "describe" on the variables of type 'object' as follows

df.describe(include=['object'])
````

### **Contagem de Valores (`value_counts`)**

A função `value_counts()` é uma ótima maneira de entender **quantas unidades existem de cada característica/variável** dentro do conjunto de dados. Podemos aplicá-la, por exemplo, à coluna `'drive-wheels'` para contar a frequência de cada categoria de **tração**.

#### **Importante:**
- O método `value_counts()` funciona **apenas** em **series do Pandas**, não em **dataframes do Pandas**.
- Isso significa que ao chamar esse método, devemos incluir **apenas um conjunto de colchetes**, e não dois.

#### **Exemplo de uso:**
```python
df['drive-wheels'].value_counts()
```

Caso seja necessário converter essa **series** em um **dataframe**, podemos fazer da seguinte maneira:

```python
df_drive_wheels = df['drive-wheels'].value_counts().to_frame()
```

Agora, `df_drive_wheels` é um dataframe contendo a contagem de cada categoria da variável `'drive-wheels'`.

````python
df['drive-wheels'].value_counts().to_frame()
````

````python
# Vamos repetir o passo anterior, mas vamos salvar o resultado em um dataframe
#"drive_wheels_counts" e renomear a coluna "drive-wheels" para "value_counts"

drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.reset_index(inplace=True)
drive_wheels_counts=drive_wheels_counts.rename(columns={'drive-wheels':'value_counts'})
drive_wheels_counts
````

### **Fundamentos do Agrupamento (`groupby`)**  

O método `groupby()` do **Pandas** é utilizado para **agrupar dados** com base em diferentes categorias. Ele permite organizar o conjunto de dados com **uma ou várias variáveis** e realizar análises sobre **grupos individuais**.

#### **Principais usos do `groupby()`:**
- **Agrupar dados** de acordo com valores categóricos
- **Resumir informações** por categoria (média, contagem, soma, etc.)
- **Aplicar funções estatísticas** para entender padrões dentro de cada grupo

#### **Exemplo básico de uso**
```python
df.groupby('categoria').mean()
```
Esse código agrupa os dados pela variável `"categoria"` e calcula a **média** de cada grupo.


````python
# Exemplo se agruparmos a variavel "drive-wheels", 
# podemos observar que são 3 diferentes categoria

df['drive-wheels'].unique()
````

Se quisermos saber, em média, qual tipo de **tração das rodas** é mais valioso, podemos agrupar a coluna `'drive-wheels'` e calcular a média dos preços.

Podemos selecionar as colunas **'drive-wheels'**, **'body-style'** e **'price'**, e então atribuí-las à variável `'df_group_one'`. Isso nos permitirá analisar a relação entre esses atributos e o preço do veículo.


````python
df_group_one = df[['drive-wheels', 'body-style', 'price']]
````

````python
df_grouped = df_group_one.groupby(['drive-wheels'], as_index=False).agg({'price':'mean'}).round(2)
df_grouped

# Para os nossos dados, parece que 'rwd' são em media mais caros, enquanto o '4ws' e 'fwd'
#tem uma media de precos aproximadas
````

````python
# Podemos agrupar tbm por multiplas variaveis. Por exemplo 'drive-wheels' e 'body-style'

df_group_one = df[['drive-wheels', 'body-style', 'price']]
df_grouped_mult = df_group_one.groupby(['drive-wheels', 'body-style'], as_index=False).mean().round(2)
df_grouped_mult
````

Estes dados agrupados são muito mais simples de se visualizarem quando apresentados como uma
pivot table. With one variable along the column and another along the row.

Nos podemos converter o dataframe em uma pivot table usando o metodo 'pivot' para criar uma
pivot table dos grupos

````python
grouped_pivot = df_grouped_mult.pivot(index='drive-wheels', columns='body-style')
grouped_pivot
````

````python
# em grande parte das vezes nao temos dados em algumas celulas da tabela pivot
# podemos entao preencher essas ceulas vazias com o vslor 0

grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
````

### **Correlação e Causalidade**  

**Correlação:** Uma medida do grau de interdependência entre variáveis.  
**Causalidade:** A relação entre **causa e efeito** entre duas variáveis.  

Embora a correlação possa sugerir que duas variáveis estão **relacionadas**, isso **não significa necessariamente que uma causa a outra**.

### **Correlação de Pearson**
- Mede a **dependência linear** entre duas variáveis **X e Y**.
- O **coeficiente de correlação** resultante varia entre **-1 e 1**.
  - **+1** → Correlação positiva forte (quando uma variável aumenta, a outra também aumenta).
  - **-1** → Correlação negativa forte (quando uma variável aumenta, a outra diminui).
  - **0** → Nenhuma correlação significativa entre as variáveis.

**Pearson Correlation** é o **método padrão** da função `'corr'` em **Pandas** para calcular correlações.


````python
df.select_dtypes(include=['number']).corr()
````

### **Valor P (P-value)**  

O **valor P** é a **probabilidade** de que a correlação entre duas variáveis seja **estatisticamente significativa**. Ele ajuda a determinar **se uma relação observada nos dados é real ou apenas fruto do acaso**.

#### **Interpretação do Valor P:**
- Um valor **P < 0.05** indica que há **menos de 5% de chance** de que a correlação seja devido ao acaso, ou seja, **existe uma relação significativa entre as variáveis**.
- Um valor **P > 0.05** sugere que a relação entre as variáveis **não é estatisticamente significativa**, indicando que pode ser apenas ruído nos dados.

Normalmente, escolhemos um **nível de significância** de **0.05**, o que significa que estamos **95% confiantes** de que a correlação entre as variáveis é significativa.



````python
from scipy import stats
````

````python
# Wheel_Base vs. Price

pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The pearson Correlation is", pearson_coef, " with a P-value of P=", p_value)

#Conclusion
#Since the p-value is < 0.001, the correlation between wheel-base and price is 
# statisically significant, although the linear relationship isnt extremely 
# strong (~0.585)
````

````python
# Horsepower vs. Price
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The pearson Correlation is", pearson_coef, " with a P-value of P=", p_value)

#Conclusion
#Since the p-value is < 0.001, the correlation between horsepower and price is 
# statisically significant, and the linear relationship is quite
# strong (~0.809, close to 1)
````
