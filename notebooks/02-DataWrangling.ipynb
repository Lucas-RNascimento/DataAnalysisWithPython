### Deal with the missing data

<b>How should you deal with missing data?</b>

<ol>
    <li>Drop data<br>
        a. Drop the whole row<br>
        b. Drop the whole column
    </li>
    <li>Replace data<br>
        a. Replace it by mean<br>
        b. Replace it by frequency<br>
        c. Replace it base on other functions
    </li>   
</ol>


- You should only drop whole columns if most entries in the column are empty. In the data set, none of the columns are empty enough to drop entirely.

- You have some freedom in choosing which method to replace data; however, some methos may seem more reasonable than others. Apply each method to different columns


<b>Replace by mean:</b>
<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke":  4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>
<ul>
    <li>"num-of-doors": 2 missgin data, replace them with "four"</li>
        <ul>
            <li>Reason: 84% sedans are four doors. Since four doors is most frquent, it is most likely to occur</li>
        </ul>
</ul>

<b>Drop the whole row:</b>
<ul>
    <li>"price": 4 missing data, delete the whole row</li>
        <ul>
            <li>Reason: You want to predict price. You cannot use any data entry without price data for prediction; therefore any row now without price data is not useful to you</li>
        </ul>
</ul>

```python
# Calculate Mean Values

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
avg_bore = df["bore"].astype("float").mean(axis=0)
avg_stroke = df["stroke"].astype("float").mean(axis=0)
avg_horsepower = df["horsepower"].astype("float").mean(axis=0)
avg_peakrpm = df["peak-rpm"].astype("float").mean(axis=0)

print("Average of normalized-losses:",avg_norm_loss)
print("Average of bore:",avg_bore)
print("Average of stroke:",avg_stroke)
print("Average of horsepower:",avg_horsepower)
print("Average of peak-rpm:",avg_peakrpm)
```

```python
# Replace "NaN" with mean values

df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avg_norm_loss)
df["bore"] = df["bore"].replace(np.nan, avg_bore)
df["stroke"] = df["stroke"].replace(np.nan, avg_stroke)
df["horsepower"] = df["horsepower"].replace(np.nan, avg_horsepower)
df["peak-rpm"] = df["peak-rpm"].replace(np.nan, avg_peakrpm)
```

```python
#Vaericando os dados apresentados em uma coluna especifica
df['num-of-doors'].value_counts()
```

```python
# verificando o dado mais comum
df['num-of-doors'].value_counts().idxmax()
```

```python
# replace the missing "num-of-doors" values by the most frequent
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, "four")
```

```python
# drop rows with NaN in "price" column
df = df.dropna(subset=["price"], axis=0)

# reset index, because we dropped two rows
df = df.reset_index(drop=True)
```

```python
# Vendo o resultado final
df.head(15)
```

<h4>What is standardization</h4>
Standardization is the process of transforming data into a common format, allowing the researcher to make the meaningfull comparison

<h4>Example</h4>
Transform mpg to L/100km:

In your data set, the fuel consuptioncolumns "city-mpg" and "highway-mpg" are represented by mpg(miles per gallon) unit. Assum you are develpping an application in a country that accespts the fuel consuption with L/100km standard.

The conversion will take this formula
L/100Km = 235 / mpg

### Data Normalization

<h4>Why normalization</h4>
Normalization is the process of transforming values of several variables into a similar range. Typical normalizarions include

1. scaling the variable so the variable average is 0
2. scaling the variable so the variance is 1
3. scaling the variable so the variable values range from 0 to 1

<h4>Example</h4>
To demonstrate normalization, say you want to scale the columns "length", "width" and "height"

Target: normalize those variables so their value ranges from 0 to 1
Approach: replace the original value by (original value)/(maximum value)

```python
# replace (original value) by (original value)/(maximum value)
df['lenght'] = df['lenght']/df['lenght'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df['height'] = df['height']/df['height'].max()
```

```python
# show the scaled columns
df[['lenght','width','height' ]]
```

### Binning
<h4>Why binning?</h4>
Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis


<h4>Example</h4>
In your data set, "horsepower" is a real valued variable ranging from 48 to 288 and it has 59 unique values. What if you only care about the price difference between cars with high horsepower, medium hoserpower, and little horsepower (3 types?) You can rearrenge them into 3 'bins' to simplify analysis

Use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins

```python
# Convert data to correct format

df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
# Plot the histogram of horsepower to see the distribution of horsepower
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y lables and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
plt.pyplot.show()
```

```python
#Find 3 bins of equal size bandwidth by using Numpy's linspace(start_value, end_value, numbers_generated) function

#Since you want to include the minimum value of horsepower, set start_value = min(df["horsepower"])

#Since you want to include the maximun value of horsepower, set end_value = max(df["horsepower"])

#Since you are building 3 bins of equal lenght, you need 4 dividers, so numbers_generated = 4

bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]), 4)
bins
```

```python
# Set Group Names
group_names = ['Low', 'Medium', 'High'] 
```

```python
#Apply the function"cut" to determine what each value of df['horsepower'] belongs to
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
# See the number of vehicles in each bin
df['horsepower-binned'].value_counts()
```

```python
#Plot distribuition each bin
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y lables and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
plt.pyplot.show()
```

```python
# Bins Visualizarion

import matplotlib as plt
from matplotlib import pyplot

#draw histogram of attribute "horsepower" with bins=3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y lables and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
plt.pyplot.show()
```

### Indicator Value

What is an indicator variable?

You use indicator variables so you can use categorical variables for regression analysis in the later modules

Example The column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, you can convert "fuel-type" to indicator variables

```python
df.columns
```

```python
# Get the indicator variables and assing it to data frame "dummy_variable_1":

dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#Change the column names for clarity
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
#merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis=1)

#drop original column "fuel-type" from "df"
df.drop("fuel-type", axis=1, inplace=True)
```

```python
df.head()
```

```python
# Doing Similar to "aspiration" variable
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.head()
```

```python
#merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_2], axis=1)

#drop original column "fuel-type" from "df"
df.drop("aspiration", axis=1, inplace=True)
df.head()
```
